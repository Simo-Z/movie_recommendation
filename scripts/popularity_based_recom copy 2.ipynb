# Movie Recommender - Popularity based

```python
import pandas as pd
import seaborn as sns
```

```python
path = "../ml-latest-small/"
movies = pd.read_csv(f"{path}movies.csv")
links = pd.read_csv(f"{path}links.csv")
ratings = pd.read_csv(f"{path}ratings.csv")
```

```python
ratings_dense_matrix = (
    ratings
        .drop(columns="timestamp")
        .groupby("movieId")
        .agg(mean_rating= ("rating", "mean"), count_rating= ("rating","count"))
        .assign(ratingtest = lambda x: x["mean_rating"] * (x["count_rating"] / 300))
        .round(1)
        
)
ratings_dense_matrix
```

## Recommender Function
**Input:** 
- ‘n’ (n is an integer, how many movies we wish to display)

**Output:** 
-  top n movies, so that you can populate the first row of WBSFLIX. Be careful not to recommend
any 5 star movies with only one rating.

```python
def top_n_movies_chat_bot_popularity_based():
    n_movies= int(input("Hey I am a basic Movie recommender Bot. I can show you a list of the most popular Movies. \n How many movies should I show you?"))
    print(f"This are the top {n_movies} Movies:")
    return  (
        ratings_dense_matrix
            .sort_values("ratingtest", ascending=False)
            .query("count_rating > 1")
            .head(n_movies)
            .reset_index()
            .merge(movies)
            .filter(["title", "mean_rating","count_rating", "ratingtest"])
    )

top_n_movies_chat_bot_popularity_based()
```
